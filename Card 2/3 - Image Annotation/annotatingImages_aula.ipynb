### Dentro desse notebook vai ser feito anotações dentro da imagem:
#### Desenho de linhas 
#### Desenho de círculos 
#### Desenho de retangulos 
#### Adicionar texto 

```python
#importação das bibliotecas
import os
import cv2
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)

%matplotlib inline
```

```python
#Função passsada no curso para baixar o que vai ser utilizado durante a aula
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    urlretrieve(url, save_path)

    try:
        with ZipFile(save_path) as z:
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
        
URL = r"https://www.dropbox.com/s/48hboi1m4crv1tl/opencv_bootcamp_assets_NB3.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB3.zip")

if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path) 
```

```python
#lendo a imagem
image = cv2.imread("Apollo_11_Launch.jpg", cv2.IMREAD_COLOR)

#plotando a imagem em escala de cinza para fazer as anotações coloridas
plt.imshow(image[:, :, ::-1])
```

```python
#Copiando a imagem para podermos desenhar uma linha nela
imageLine = image.copy()

#Os 2 primeiros parametros é onde a linha começa e termina 
#O 3 parametro é a cor da linha em BGR
#Grossura da linha é 5 pixels
#O tipo da linha é anti-alias, que utiliza pixels semi transparentes

cv2.line(imageLine, (200, 100), (400, 100), (0, 255, 255), thickness=5, lineType=cv2.LINE_AA);

#plotando a imagem com a linha
plt.imshow(imageLine[:,:,::-1])
```

```python
#Copiando a imagem para desenhar um círculo
imageCircle = image.copy()

#A diferença agora é que não é passado onde começa e onde termina. Agora é passado o centro e o raio do círculo
cv2.circle(imageCircle, (900,500), 100, (0, 0, 255), thickness=-2, lineType=cv2.LINE_AA);

#plotando a imagem
plt.imshow(imageCircle[:,:,::-1])
```

```python
#Desenhar um retangulo

imageRectangle = image.copy()

#Parecido com a da linha porém o 2 parametro é o começo do topo à esquerda e o terceiro o fim de baixo a direita
cv2.rectangle(imageRectangle, (500, 100), (700, 600), (255, 0, 255), thickness=-5, lineType=cv2.LINE_8)

plt.imshow(imageRectangle[:, :, ::-1])
```

```python
imageText = image.copy()

text = "Apollo 11 Saturn V Launch, July 16, 1969" #O que vai ser escrito
fontScale = 2.3 #Tamanho da fonte
fontFace = cv2.FONT_HERSHEY_PLAIN #Fonte
fontColor = (0, 255, 0) #Cor da fonte
fontThickness = 2 #Grossura da fonte
 
#O terceiro parametro é o de onde a escrita começa
cv2.putText(imageText, text, (200, 700), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA);

#plotando a imagem
plt.imshow(imageText[:, :, ::-1])
```
